```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Count")
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/filter.txt')
```

```
rdd2 = rdd.flatMap(lambda x: x.split(' '))
```

```
rdd2.collect()
```

```
rdd2.count()
```

```
rdd2.countByValue()
```

```
rdd2.countByKey()
```
