```
# Databricks notebook source
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Filter")
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/sample.txt')
```

```
rdd.collect()
```

```
def foo(x):
  if x == '85 965 257 35':
    return False
  else:
    return True
```

```
rdd2 = rdd.filter(foo)
```

```
rdd2.collect()
```
