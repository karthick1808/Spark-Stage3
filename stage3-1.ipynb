```
import pyspark
```

```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("QUIZ")
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/sampletext.txt')
```

```
rdd.collect()
```

```
def quiz(x):
  # x -> 'great'
  l = x.split(' ') # l -> ['great']
  l2 = []
  for s in l:
    l2.append(len(s))
  return l2
```

```
rdd2 = rdd.map(quiz)
```

```
rdd2.collect()
```

```
rdd3=rdd.map(lambda x: x.split(" "))
```

```
rdd3.collect()
```

```
rdd4 = rdd.map(lambda x: [len(s) for s in x.split(' ')])
```

```
rdd4.collect()
```
